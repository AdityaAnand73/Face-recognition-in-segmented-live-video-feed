## Loading the Model:

```python
from keras.models import load_model

classifier = load_model('F1.h5')
```

## Live testing:

```python
import cv2
import numpy as np
import os
import IPython.display as ipd
import webbrowser as wb

Person = { "[0]": "Aditya",
           "[1]": "Anjali",
           "[2]": "Divyansh"
          }

face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def face_detector(img):

    # Convert image to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is ():
        return (None,None,None)

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        cropped_image = img[y:y+h, x:x+w]
        cropped_image = cv2.resize(cropped_image, (224, 224))
        
    return img, cropped_image, faces


# Open Webcam
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    h, w, c = frame.shape
    img, cropped_face, face_coor = face_detector(frame[:,:int(w/2)])
    frame = cv2.line(frame, (int(w/2),0), (int(w/2),h), (0,0,255), 2)
    
    try:
        input_im = cropped_face
        input_im = cv2.resize(input_im, (224, 224), interpolation = cv2.INTER_LINEAR)
        input_im = input_im / 255.
        input_im = input_im.reshape(1,224,224,3) 
        
        res = np.argmax(classifier.predict(input_im, 1 , verbose=0),axis=1)

        if face_coor is not None:
            for (x,y,w,h) in face_coor:
                cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)

            cv2.putText(frame, "{}".format(Person[str(res)]), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        

    except:
        print("!!Face Not Found!!")
        pass
    
    cv2.imshow('Person', frame)
    
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break

cap.release()
cv2.destroyAllWindows()
```
